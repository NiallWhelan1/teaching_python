#### Import Packages & Modules

```python
import os
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

## Print plots inline (in notebook)
%matplotlib inline
```

```python
# Set aesthetic parameters in one step.
sns.set(context='notebook',
        style='darkgrid',
        palette='deep',
        font='sans-serif',
        font_scale=1,
        color_codes=True,
        rc=None,)
```

#### Data Input

```python
# Read dataset
df = pd.read_excel('Pokemon.xlsx', index_col=0)


# Display first 5 observations
df.head()
```

#### Default Scatterplot

```python
sns.lmplot(x='Attack', y='Defense', data=df, ## Required to generate plot
          fit_reg = True,
          ci = 75);


# Scatterplot arguments
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=False, # No regression line
           hue='Stage');   # Color by evolution stage 
            
```

#### Scatterplot - Improving with Matplotlib

```python
# Plot using Seaborn
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=False, 
           hue='Stage');
 
# Tweak using Matplotlib
plt.ylim(0, None); ## Setting the min on y axis to 0
plt.xlim(0, None); ## Setting the min on x axis to 0
```

#### Seaborn Themes

#### Plot Size/Area

```python
## Use matplotlib to initialise the plot area using figure
plt.figure(figsize=(20, 10));

# Violin plot
sns.violinplot(x='Type 1', y='Attack', data=df);
```

#### Create Color Pallete

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]

plt.figure(figsize=(20, 10));
# Violin plot with Pokemon color palette
sns.violinplot(x='Type 1', y='Attack', data=df, 
               palette=pkmn_type_colors); # Set color palette
```

#### Swarm Plot

```python
plt.figure(figsize=(20, 15));
# Swarm plot with Pokemon color palette
sns.swarmplot(x='Type 1', y='Attack', data=df, 
              palette=pkmn_type_colors);
```

#### Overlaying Plots (Example: Violin & Swam plot)

```python
# Set figure size with matplotlib
plt.figure(figsize=(12,8))
 
# Create plot
sns.violinplot(x='Type 1',
               y='Attack', 
               data=df, 
               inner=None, # Remove the bars inside the violins
               palette=pkmn_type_colors)
 
sns.swarmplot(x='Type 1', 
              y='Attack', 
              data=df, 
              color='k', # Make points black
              alpha=0.7) # and slightly transparent
 
# Set title with matplotlib
plt.title('Attack by Type');
```

#### Pandas Melt Data

```python
## Remove unwanted columns
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)

## Current Setup
print(stats_df.head())

print('\n\n-------\n\n')

# Melt DataFrame
melted_df = pd.melt(stats_df, 
                    id_vars=["Name", "Type 1", "Type 2"], # Variables to keep
                    var_name="Stat") # Name of melted variable

print(melted_df.head())

print('\n\n-------\n\nMelt Causes complete shape change.\n\n')

print(f"Previously: {stats_df.shape}")
print(f"Melted: {melted_df.shape}")

print('\n\n-------\n\nEach pokemon now has 6 rows - 1 for each melted attribute.\n\n')
melted_df.sort_values(by = 'Name').head(12)
```

#### Swarmplot with melted DF

```python
plt.figure(figsize=(12,8));
# Swarmplot with melted_df
sns.swarmplot(x='Stat', y='value', data=melted_df, 
              hue='Type 1');
```

```python
### Plot Cleanup ###


# 1. Enlarge the plot
plt.figure(figsize=(10,6))
 
sns.swarmplot(x='Stat', 
              y='value', 
              data=melted_df, 
              hue='Type 1', 
              split=True, # 2. Separate points by hue
              palette=pkmn_type_colors) # 3. Use Pokemon palette
 
# 4. Adjust the y-axis
plt.ylim(0, None)
 
# 5. Place legend to the right
plt.legend(bbox_to_anchor=(1, 1), loc=2);
```

#### Correlation Heatmap

```python
print(stats_df.corr())
# Heatmap
sns.heatmap(stats_df.corr(),);
```

#### Histogram (a.k.a. Distplot)

```python
# Distribution Plot (a.k.a. Histogram)
sns.distplot(df.Attack);
plt.xlim(0, None);
```

#### Barplot (a.k.a. Countplot)

```python
# Count Plot (a.k.a. Bar Plot)
sns.countplot(x='Type 1', data=df, palette=pkmn_type_colors)
 
# Rotate x-labels
plt.xticks(rotation=-45);
```

#### Categorical/Factor Plot
Categorical or Factor plots make it easy to separate plots by categorical classes.



```python
# Factor Plot
g = sns.catplot(x='Type 1', 
                   y='Attack', 
                   data=df, 
                   hue='Stage',  # Color by stage
                   col='Stage',  # Separate by stage
                   kind='swarm') # Swarmplot
 
# Rotate x-axis labels
g.set_xticklabels(rotation=-45);
 
# Doesn't work because only rotates last plot
# plt.xticks(rotation=-45)
```

#### Density Plot

```python
# Set figure size with matplotlib
plt.figure(figsize=(12,8))
 
# Density Plot
sns.kdeplot(df.Attack, df.Defense)
```

#### Joint Distribution Plot
Joint distribution plots combine information from scatter plots and histograms to give you detailed information for bi-variate distributions.

```python
# Joint Distribution Plot
sns.jointplot(x='Attack', y='Defense', data=df);
```
