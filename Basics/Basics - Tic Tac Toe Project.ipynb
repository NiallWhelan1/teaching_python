# Tic Tac Toe
2 players should be able to play the game (both sitting at the same computer)

The board should be printed out every time a player makes a move

You should be able to accept input of the player position and then place a symbol on the board

```python
from IPython.display import clear_output

def ttt_setup():

    # Board Doctionary
    global board_dict
    board_dict = dict.fromkeys(range(1,10), ' ')
    
    # Turn Counter
    global turn_counter
    turn_counter = 1
  
def board_setup():
    print("\n")
    print(f" 1 | 2 | 3")
    print(f" ---------")
    print(f" 4 | 5 | 6")
    print(f" ---------")
    print(f" 7 | 8 | 9")
    print("\n")

def board_output(board_dict):
    print("\n")
    print(f" {board_dict[1]} | {board_dict[2]} | {board_dict[3]}")
    print(f" ---------")
    print(f" {board_dict[4]} | {board_dict[5]} | {board_dict[6]}")
    print(f" ---------")
    print(f" {board_dict[7]} | {board_dict[8]} | {board_dict[9]}")
    print("\n")
    
def player_input():
    global turn_counter
    
    if turn_counter % 2 == 0:
        player_var = 2
    else:
        player_var = 1
    input_value = int(input(f'\nPlayer {player_var}, please choose a location on board:'))

    x = 0
    while x == 0: 
        if not input_value in range(1,10):
            input_value = int(input(f'\nPlayer {player_var}, please choose a valid (1-9) location on board:'))
        elif board_dict[input_value] != ' ':
            input_value = int(input(f'\nPlayer {player_var}, please choose a different location on board:'))
        else: 
            x = 1

    turn_counter += 1

    return player_var, input_value

def board_update(player_var, input_value):
    global board_dict
    
    if player_var == 1:
        mark = 'X'
    else:
        mark = 'O'
        
    board_dict[input_value] = mark

def game_status(board_dict):
    clear_output()
    if (board_dict[1] == board_dict[2] == board_dict[3]) and board_dict[1] != ' ':
        winner = board_dict[1]
    elif(board_dict[1] == board_dict[4] == board_dict[7]) and board_dict[1] != ' ':
        winner = board_dict[1]
    elif(board_dict[1] == board_dict[5] == board_dict[9]) and board_dict[1] != ' ':
        winner = board_dict[1]
    elif(board_dict[2] == board_dict[5] == board_dict[8]) and board_dict[2] != ' ':
        winner = board_dict[2]
    elif(board_dict[3] == board_dict[6] == board_dict[9]) and board_dict[3] != ' ':
        winner = board_dict[3]
    elif(board_dict[3] == board_dict[5] == board_dict[7]) and board_dict[3] != ' ':
        winner = board_dict[3]
    elif(board_dict[4] == board_dict[5] == board_dict[6]) and board_dict[4] != ' ':
        winner = board_dict[4]
    elif(board_dict[7] == board_dict[8] == board_dict[9]) and board_dict[7] != ' ':
        winner = board_dict[7]
    else:
        winner = None
        
    if winner == 'X':
        board_output(board_dict)
        print(f"\nPlayer 1 has won the game. Congrats.")
        return 1
    elif winner == 'O':
        print(f"\nPlayer 2 has won the game. Congrats.")
        board_output(board_dict)
        return 2
    else:
        board_output(board_dict)
        return None

    
def new_game():
    new_game = input('Would you like to play another game (Yes or No):')
    while not new_game.lower() in ['yes', 'no']:
        new_game = input('Invalid Choice. Would you like to play another game (Yes or No):')
    
    if new_game.lower() == 'yes':
        clear_output()
        '\n\nGreat! New Game. Best of luck.\n\n'
        game_func()
    else:
        print('\n\nGame over. Thanks for playing.')
        
    
def game_func():
    ttt_setup()
    
    print('Welcome to Tik Tac Toe.')
    
    board_setup()
    
    winner = None
    while winner == None:
        player_var, input_value = player_input()
    
        board_update(player_var, input_value)
        
        winner = game_status(board_dict)
        
        if turn_counter > 9:
            print('\nThe game is over, no winner')
            break

    new_game()
            
            
```
